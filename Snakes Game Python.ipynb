```python
import turtle
import time
import random

delay = 0.1
s=0
h=0

wn=turtle.Screen()
wn.title("Snakes")
wn.bgcolor("black")
wn.setup(width=600,height=600)
wn.tracer(0)

title=turtle.Turtle()
title.speed(0)
title.penup()
title.hideturtle()
title.color('white')
title.goto(0,40)

title2=turtle.Turtle()
title2.speed(0)
title2.penup()
title2.hideturtle()
title2.color('white')
title2.goto(0,-40)

title.write(f"Welcome to Snakes",align="center",font=("Courier",24,"normal"))
title2.write(f"Created by Amogh. press Enter :)",align="center",font=("Courier",24,"normal"))

time.sleep(1)
wn.listen()
def clr():
    title.clear()
    title2.clear()

wn.onkeypress(clr,'Return')

food=turtle.Turtle()
food.speed(0)
food.shape('circle')
food.color('red')
food.penup()
food.goto(0,100)

top=turtle.Turtle()
top.speed(0)
top.shape("square")
top.color("green")
top.penup()
top.goto(0,0)
top.direction='stop'

score=turtle.Turtle()
score.speed(0)
score.color('white')
score.penup()
score.hideturtle()
score.goto(0,260)


body=[]

def up():
    top.direction='up'
    
def down():
    top.direction='down'
    
def left():
    top.direction='left'
    
def right():
    top.direction='right'
    
wn.listen()
wn.onkeypress(up, 'w')
wn.onkeypress(down,'s')
wn.onkeypress(left,'a')
wn.onkeypress(right,'d')

def move():
    if top.direction=='up':
        y=top.ycor()
        top.sety(y+20)
    if top.direction=='down':
        y=top.ycor()
        top.sety(y-20)
    if top.direction=='left':
        x=top.xcor()
        top.setx(x-20)
    if top.direction=='right':
        x=top.xcor()
        top.setx(x+20)
         
         

while True:
    
    wn.update()
    
    if top.distance(food)<20 :
        x=random.randint(-290,290)
        y=random.randint(-290,290)
        food.goto(x,y)
        
        add=turtle.Turtle()
        add.speed(0)
        add.shape('square')
        add.color('green')
        add.penup()
        body.append(add)
        s+=1
        if s>h:
            h=s
      
        
    for index in range (len(body)-1,0,-1):
        x=body[index-1].xcor()
        y=body[index-1].ycor()
        body[index].goto(x,y)
        
        
    
    if len(body)>0:
        x=top.xcor()
        y=top.ycor()
        body[0].goto(x,y)
    

        
    if top.xcor()==300 or top.xcor()==-300 or top.ycor()==300 or top.ycor()==-300:
        time.sleep(1)
        top.goto(0,0)
        s=0
        top.direction="stop"
        for x in body:
            x.goto(1000,1000)
            
        body.clear()
    
    score.clear()
    score.write(f"Score={s} Highscore={h}",align="center",font=("Courier",24,"normal"))
    move()
    time.sleep(delay)

wn.mainloop()
```
